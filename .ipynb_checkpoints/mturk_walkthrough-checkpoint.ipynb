PsiTurk is a platform that allows us to launch our experiments to MTurk in a clean, efficient manner. PsiTurk has been developed by several different teams over the years, but most recently has been led by the Guericks Lab at NYU.

<h3><center>Heroku</center></h3>

Heroku is a platform that allows developers (or researchers) to build and run applications (i.e. experiments) entirely on the cloud. Without Heroku, we would have to run our experiments on our local machines, this would mean launching the server, and anytime we shut off our computer the server would shut down as well. Instead the server can run at all times, allowing for data collection at anytime throughout the day. Heroku also offers nice compatibility with POSTGRES databases, which makes our lives quite easy when it comes to storing and backing up our data. 

Once again the first step you will need to take in setting up heroku is creating an account, which you can do here: https://signup.heroku.com/. Once you have an account make sure to download the Heroku CLI, instructions here: https://devcenter.heroku.com/articles/heroku-cli.

```python
## First let's import the libraries we would like to use
from sqlalchemy import create_engine
import pandas as pd
import subprocess
## connect to db
DATABASE_URL = subprocess.check_output("heroku config:get DATABASE_URL --app spacetop", shell=True).decode('utf-8')
engine = create_engine(DATABASE_URL)
## query results in participants table
result = pd.read_sql_query(
    '''
SELECT * FROM participants
    ''',
    con=engine)
result.head(5)
```

```python
##now store our column of data, get rid of none values, and convert to dict
datastr = result['datastring']
meta = datastr.dropna()
meta.to_dict()
meta
```
